# Logistic regression. Using the data provided (Logistic_regression_ls.csv), plot the decision boundary (linear) using Optimization of the sigmoid function.

Logistic_regression_ls.csv Logistic_regression_ls.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Logistic_regression_ls.csv')

# Assuming the dataset has columns 'X1', 'X2', and 'label' (binary 0/1)
X = df[['x1', 'x2']].values
y = df['label'].values

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Initialize parameters (weights and bias)
theta = np.zeros(X.shape[1] + 1)  # Additional dimension for the bias term

learning_rate = 0.01
num_iterations = 20000

# Gradient descent optimization
for _ in range(num_iterations):
    z = np.dot(X, theta[1:]) + theta[0]  # Linear combination of features and parameters
    predictions = sigmoid(z)  # Compute the predictions 

    # Compute the gradient and update parameters
    gradient = np.dot(X.T, predictions - y) / len(y)
    theta[1:] -= learning_rate * gradient  # Update weights
    theta[0] -= learning_rate * np.sum(predictions - y) / len(y)  # Update bias
    
print("Final updated weights",theta)
# Plot the decision boundary
x1_values = np.linspace(np.min(X[:, 0]), np.max(X[:, 0]), 100)
x2_values = -(theta[0] + theta[1] * x1_values) / theta[2]  # Decision boundary (linear)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.plot(x1_values, x2_values, color='red', label='Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Logistic Regression Decision Boundary')
plt.legend()
plt.show()
```

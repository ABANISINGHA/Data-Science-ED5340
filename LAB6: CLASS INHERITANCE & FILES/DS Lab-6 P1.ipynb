# 1) Design a class named Polygon that initializes with the length of a side. Then, derive a class named Square from the Polygon class. Utilize the side length defined in the Polygon class for the Square class. Within the Square class, implement a method called findArea() that calculates and returns the square's area based on its side length. Use __init__() for necessary initialization

```python
class Polygon:
    def __init__(self,l):
        self.len_of_side = l
    def display_polygon(self):
        print("The length of the sides of the polygon is:",self.len_of_side,'unit')
class Square(Polygon):
    def __init__(self,l,n=4):
        super().__init__(l)
        self.num_of_sides = n
    def findArea(self):
        return self.len_of_side**2
p1 = Polygon(3)
p1.display_polygon()
s1 = Square(3)
A = s1.findArea()
print(f"The area of the square is: {A} unit square")
```

```python
class Father:
    def __init__(self,f_name,f_age,f_talents):
        self.father_name = f_name
        self.father_age = f_age
        self.father_talents = f_talents
    def father_details(self):
        print(f"father's name {self.father_name},father's age {self.father_age} year,father's talents are {self.father_talents}")

        
class Mother:
    def __init__(self,m_name,m_age,m_talents):
        self.mother_name = m_name
        self.mother_age = m_age
        self.mother_talents = m_talents
    def mother_details(self):
        print(f"mother's name {self.mother_name},mother's age {self.mother_age} year,mother's talents are {self.mother_talents}")

class Child(Father,Mother):
    def __init__(self, f_name = '', f_age=0, f_talents='', m_name='', m_age=0, m_talents='', c_name='', c_age=0, c_gender=''):
        Father.__init__(self,f_name,f_age,f_talents)
        Mother.__init__(self,m_name,m_age,m_talents)
        child_name = c_name
        child_age = c_age
        child_gender = c_gender
    
    def getChildDetails(self):
        self.child_name = input("Enter the name of the child: ")
        self.child_age = int(input("Enter the age of the child: "))
        self.child_gender = input("Enter the age of the child: ")
        self.father_name = input("Enter the name of the father: ")
        self.father_age = int(input("Enter the age of the father: "))
        self.father_talents = list(input("Enter the talents of the father: ").split(","))
        self.mother_name = input("Enter the name of the mother: ")
        self.mother_age = int(input("Enter the age of the mother: "))
        self.mother_talents = list(input("Enter the talents of the mother: ").split(","))
        
    def printChildDetails(self): 
        print(f"Child's name {self.child_name},child's age {self.child_age} year,child's gender {self.child_gender}")
        Father.father_details(self)
        Mother.mother_details(self)
        
    def displayTalents(self):
        child_talent = []
        for talent in list(self.father_talents):
            if talent in list(self.mother_talents):
                child_talent.append(talent)
        if child_talent == []:
            print("No talent is inherited from father and mother.")
        else:
            print("Child's talents are: ",child_talent)

        
    
        
f = Father("Rabi Kumar",45,"Singing,Teaching")
f.father_details()
child = Child()
child.getChildDetails()
child.printChildDetails()
child.displayTalents()
```

# 3.  Text File Input Output

         Create a .txt (text) file and use the pledge of India as the content of the text file.

         Write a python program that reads this text file, processes it by counting the number of occurrences of each word in the file, and then writes the result back to a          new text file.



```python
def count_word_occurrences(file_path):
    word_count = {} # taking an empty dictonary
    with open(file_path, 'r') as file:
        for line in file:
            words = line.split()
            for word in words:
                word = word.strip().lower().rstrip('.,?!')   # Removing punctuation and converting to lowercase
                if word:
                    if word in word_count:
                        word_count[word] += 1
                    else:
                        word_count[word] = 1
    return word_count

def write_word_occurrences(word_count, output_file):
    with open(output_file, 'w') as file:
        for word, count in word_count.items():
            file.write(f"{word}: {count}\n")


input_file = "india_pledge.txt"
output_file = "word_occurrences.txt"

word_count = count_word_occurrences(input_file)

write_word_occurrences(word_count, output_file)

print("Word occurrences counted from india_pledge.txt and written to word_occurrences.txt", output_file)
```

```python
class RestaurantBill(Bill):
    def __init__(self,name = '',tab_no = 0,order = '',Price_List="Item1-100,Item2-70,Item3-250"):
        super().__init__(name,tab_no,order)
        self.Price_List = Price_List
        self.service_charge = 50  # Taken over the order of customer
        self.tax = 10    # Tax is taken as 10% per item
        
    def str_to_Pricelst(self):
        lst = list(self.Price_List.split(','))
        Price_lst = []
        for str in lst:
            ls = []
            for st in list(str.split("-")):
                ls.append(st)
            Price_lst.append(ls)
        return Price_lst
    def make_bill(self):
        total_bill = 0
        order_lst = super().str_to_lst()
        price_lst = self.str_to_Pricelst()
        for i in range(len(order_lst)):
            for j in range(len(order_lst[0])):
                if order_lst[i][j] == price_lst[i][j]:
                    total_bill = total_bill + int(order_lst[i][1])*int(price_lst[i][1])*(1 + self.tax/100)             
        return total_bill + self.service_charge
    def Order_details_with_bill(self):
        #super().order_details()
        print("Name of the customer:", self.Customer_name)
        print("Table number: ",self.Table_Number)
        for i in super().str_to_lst():
            print(f"{i[0]} : {i[1]} piece","\n")
        print("Price of the items:")
        for i in self.str_to_Pricelst():
            print(f"{i[0]} : ${i[1]}","\n")
        print("Total bill: ",self.make_bill(),f"(Service charge ${self.service_charge} and tax {self.tax}% per item included.)")

        
resturant_bill_1 = RestaurantBill("Akshay",7,"Item1x1,Item2x3,Item3x1","Item1-100,Item2-70,Item3-250")
resturant_bill_1.Order_details_with_bill()
```

```python
msg2= ' name : Akshay'
msg3 = '\n  table number : 7 '
msg4 = '\n order detail : Item1x1,Item2x3,Item3x1 '

f=open('customerdetails.txt','w')
f.write(msg2)
f.write(msg3)
f.write(msg4)

f.close()




f=open('customerdetails.txt','r')
data=f.read()
print(data)






msg6= '{"Item1": 100, "Item2": 70, "Item3": 250}'

f=open('pricelist.txt','w')
f.write(msg6)


f.close()

f=open('pricelist.txt','r')
data=f.read()
print(data)





class Bill:
    def __init__(self, cname, table_no, order):
        self.cname = cname
        self.table_no = table_no
        self.order = order

    def getdetails(self):
        ord = []
        sd = self.order.split(',')
        for df in sd:
            item, quantity = df.split('x')
            ord.append([item.strip(), int(quantity)])

        return ord

class Restaurant_bill(Bill):
    def __init__(self, cname, table_no, order, price_list):
        super().__init__(cname, table_no, order)
        self.price_list = price_list

    def calculate_bill(self):
        total_bill = 0
        order_details = self.getdetails()
        for item, quantity in order_details:
            if item in self.price_list:
                total_bill += self.price_list[item] * quantity

        return total_bill

    def complete_bill(self, tax=0.1, service=0.05):
        self.tax = tax
        self.service = service

        with open('complete_bill.txt', 'w') as file:
            file.write("Customer name: {}\n".format(self.cname))
            file.write("Table number: {}\n".format(self.table_no))
            file.write("Order details:\n")
            order_details = self.getdetails()
            for item, quantity in order_details:
                file.write("- {}: {}\n".format(item, quantity))
            bill_amount = self.calculate_bill()
            file.write("Bill amount: {}\n".format(bill_amount))
            file.write("Tax charge: {}\n".format(bill_amount * tax))
            file.write("Service charge: {}\n".format(bill_amount * service))
            grand_total = bill_amount + bill_amount * tax + bill_amount * service
            file.write("Grand total: {}\n".format(grand_total))


# Read customer details from file
with open('customerdetails.txt', 'r') as file:
    customer_data = file.read().splitlines()
cname = customer_data[0].split(': ')[1].strip()
table_no = int(customer_data[1].split(': ')[1].strip())
order = customer_data[2].split(': ')[1].strip()

# Read price list from file
with open('pricelist.txt', 'r') as file:
    price_list_str = file.read()
price_list = eval(price_list_str)

# Create RestaurantBill object
bill = Restaurant_bill(cname, table_no, order, price_list)

# Print complete bill to new file
bill.complete_bill()

f=open('complete_bill.txt','r')
data=f.read()
print(data)
```

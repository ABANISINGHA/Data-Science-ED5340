# 5. WAP to plot a 3-d graph of the helical wave signal using the scatter method and normal line method. Plot them separately and specify legend.



```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the helical wave signal
r = 2  # radius of the helix  
t = np.linspace(0, 4*np.pi, 100)

# The equation of the helix
x =  r*np.cos(t)
y =  r*np.sin(t)
z = t

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x, y, z , label='Scatter Method')

# Create a normal line plot
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
line = ax2.plot(x, y, z, label='Normal Line Method')

# Add legends
fig.legend()
fig2.legend()

plt.show()
```

# 10. Use the central difference method to find the first and second order derivatives of the function. Use the following function for testing the result. And also verify the result manually (Write on paper and upload jpg). Refer to section 2.5.1 of “Optimization for Engineering Design: Algorithms and Examples” by KALYANMOY DEB, 2nd edition


f(x) = 3x**2 + 2x



```python
def f(x):
    return 3*(x**2) + 2*x

# Defining the 1st and 2nd order derivative of f by using central difference scheme
def first_der_f(x,h):
    der_f = (f(x+h)-f(x-h))/(2*h)
    return der_f

def second_der_f(x,h):
    der_f = (f(x-h) - 2*f(x) + f(x+h))/(h**2)
    return der_f

#Let us consider the point x at which we are finding derivative and the step length h

x = 3
h = 0.1
print("First derivative of f by using central difference scheme:",first_der_f(x,h))
print("Second derivative of f by using central difference scheme:",second_der_f(x,h))
```
